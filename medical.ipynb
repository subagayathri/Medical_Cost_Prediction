```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("insurance.csv")
df.head(10)
```

```python
df.dtypes
```

```python
#One Hot Encoding

tem=pd.get_dummies(df["region"])
tem.head(10)
```

```python
df.drop("region",axis=1,inplace=True)
df.head(10)
```

```python
#comibing both dataframe

df=pd.concat([df,tem],axis=1)
df.head(10)
```

```python
#converting categorized variables into numeric features

map={"yes":1,"no":0}
df["smoker"]=df["smoker"].map(map)
map1={"female":0,"male":1}
df["sex"]=df["sex"].map(map1)
df.head(10)

```

```python
df.corr()
```

```python
#correlation Heatmap to find out highly correlated columns

plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm",linewidth=2)
```

```python
#scatter plot of target based on smoker

x=df["smoker"]
y=df["charges"]
plt.figure(figsize=(12,9))
plt.scatter(x,y)
plt.xlabel("Non smoker vs smoker")
plt.ylabel("charges")
```

```python
#It is clearly visible that smokers have high mean cost(13000-60000+) when compare to non smokers(0-38000)
```

```python
y=df["charges"]
x=df.drop("charges",axis=1)
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
#spliting data into 85% of train and 15% as test

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.15)
```

```python
from sklearn.linear_model import LinearRegression
#training multiplr linera regression model

reg=LinearRegression().fit(x_train,y_train)
y_pred=reg.predict(x_test)

```

```python
from sklearn.metrics import r2_score
#checking the r squared error on test data
 
r2_score(y_test,y_pred)

#storing independant variable in temp variable
p_x=x 
```

```python
from sklearn.preprocessing import PolynomialFeatures
#cahnging the data to 3rd degree polynomial

pol=PolynomialFeatures(degree=1)
p_x=pol.fit_transform(x)
p_x
```

```python
#traing model similarly but with 3rd degree polynomial of x this time

x_train,x_test,y_train,y_test=train_test_split(p_x,y,random_state=1,test_size=0.15)
reg=LinearRegression().fit(x_train,y_train)
y_pred=reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
#cross validitaing score to check and avoid ovefitting

from sklearn.model_selection import cross_val_score
c=cross_val_score(reg,p_x,y,cv=4)
c
```

```python
#Final mean accuracy
print("Mean Accuracy after cross validation is:",c.mean()*100,end="%")
```
